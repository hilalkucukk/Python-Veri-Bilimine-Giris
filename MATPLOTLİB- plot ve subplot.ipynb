```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x=["1.ABD","2.ÇİN","3.JAPONYA","4.ALMANYA","5.İNGİLTİRE","6.FRANSA"]
y=[20410,14090,5170,4210,2940,2930]
z=[63039,9957,40649,51158,44162,44915]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"r")
```

```python
plt.plot(x,y,"r")
plt.xlabel("ÜLKELER")
plt.ylabel("MİLLİ GELİR")
plt.title("ÜLKELERE GÖRE MİLLİ GELİR")
plt.show() #başka uygulamalarda göstermek için
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.subplot(1,2,2)
plt.plot(x,y,"y")
plt.tight_layout() #düzgün gözükmesi için
```

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.7,0.7])
```

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
f=plt.figure()
axes1=f.add_axes([0.1,0.1,0.9,0.9])
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,"y")
axes2.plot(x,z,"r")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,"b")
axes[0].set_title("Milli Gelir")

axes[1].plot(x,z,"r")
axes[1].set_title("Kişi Başına")

plt.tight_layout()
```
