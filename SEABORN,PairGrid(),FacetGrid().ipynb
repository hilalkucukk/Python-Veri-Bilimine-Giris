```python
import numpy as np
import pandas as pd
import matplotlib as npl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
n=pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
n
```

```python
m=sns.PairGrid(n)
m.map_diag(sns.distplot)
m.map_upper(plt.scatter)
m.map_lower(sns.kdeplot)
```

# FacetGrid

```python
s=pd.read_csv("evli.csv")
```

```python
ss=sns.FacetGrid(data=s,col="ay",row="evli")
ss.map(sns.distplot,"gelir")
```

```python
ss=sns.FacetGrid(data=s,col="ay",row="evli")
ss.map(plt.hist,"gelir")
```
