```python
n=pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
n
```

```python
sns.distplot(n["Doğum oranı"],bins=20)
```

```python
sns.set(style="dark",palette="muted",font_scale=1.5)
sns.distplot(n["Doğum oranı"],bins=20,kde=False,color="y")
plt.tight_layout()
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1)
sns.distplot(n["Yıllık değişim oranı"],hist=False,color="r",rug=True,)
plt.tight_layout()
```

```python
sns.set(style="white",palette="Blues",font_scale=1.5)
sns.distplot(n["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True})
plt.tight_layout()
```

```python
sns.set(style="whitegrid",palette="Blues",font_scale=1.5)
sns.distplot(n["Nüfüs"],color="m")
plt.tight_layout()
```

```python
sns.set(style="darkgrid",font_scale=1.5)

f, axes=plt.subplots(2,2,figsize=(15,10))

sns.distplot(n["Doğum oranı"],bins=20,kde=False,color="y",ax=axes[0,0])
sns.distplot(n["Yıllık değişim oranı"],hist=False,color="r",rug=True,ax=axes[0,1])
sns.distplot(n["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True},ax=axes[1,0])
sns.distplot(n["Nüfüs"],color="m",ax=axes[1,1])
```

# jointplot()

```python
sns.jointplot(n["Doğum oranı"],n["Ortalama yaş"],data=n)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="hex",color="r")
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="reg",xlim=(-2,6),ylim=(0,10),color="r",size=10)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="kde",xlim=(-2,6),ylim=(0,10),color="r",size=8)
```
