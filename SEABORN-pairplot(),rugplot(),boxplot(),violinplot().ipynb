```python
import numpy as np
import pandas as pd
import matplotlib as npl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
e=pd.read_csv("50ülke.csv")
e.pop("2023")
e.pop("SIRALAMA")
e.pop("SIRALAMA.2")
e
```

```python
sns.pairplot(e,hue="KITA",palette="inferno")
```

# rugplot

```python
sns.rugplot(e["BÜYÜME"],color="y",height=0.2)
sns.kdeplot(e["BÜYÜME"],color="r")
```

# boxplot()

```python
sns.boxplot(x="KITA",y="KİŞİ BAŞINA",data=e,width=0.5)
plt.tight_layout()
```

```python
sns.boxplot(x="KITA",y="BÜYÜME",data=e,palette="Set3")
```

```python
ev=pd.read_csv("evli.csv")
```

# violinplot()

```python
sns.set(style="whitegrid")
sns.violinplot(x="ay",y="gelir",data=ev,hue="evli",split=True)
```
